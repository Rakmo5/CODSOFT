```python
import re

def clean_text(text):
    
    """
    Return the cleaned version of text.
    Removes Puntuations, unwanted symbols and casing
    """
    # lowercase
    text = text.lower()
    # Removes puntuation, special chars, digits
    text = re.sub(r"[^a-zA-Z\s]","",text)
    # Remove extra whitespace
    text = re.sub(r"\s+"," ",text).strip()
    return text

# Apply it to our dataframe
train_df["clean_plot"] = train_df["plot"].apply(clean_text)
test_df["clean_plot"] = test_df["plot"].apply(clean_text)
test_soln_df["clean_plot"] = test_soln_df["plot"].apply(clean_text)
```

```python
# Most Common Words Overall (after cleaning)
from collections import Counter

# Combine all plots
all_words = " ".join(train_df["clean_plot"]).split()
common_words = Counter(all_words).most_common(20)

# Convert to dataframe for ploting
words_train_df = pd.DataFrame(common_words,columns=["word","count"])

sns.barplot(data=words_train_df,x="count",y="word")
plt.title("Top 20 Most Frequent words in Plots")
plt.show()
```

```python
# Pick a few genres like 'drama','thriller',etc.

top_genres = train_df['genre'].value_counts().head(3).index

for genre in top_genres:
    subset = train_df[train_df["genre"]==genre]
    words = " ".join(subset["clean_plot"]).split()
    common = Counter(words).most_common(10)
    word_train_df = pd.DataFrame(common,columns=["word","count"])

    plt.figure(figsize=(6,3))
    sns.barplot(data=words_train_df,x="count",y="word")
    plt.title(f"Top Words in {genre} Genre")
    plt.show()
```

```python
# Step 1: Extract the year from the original title (before modifying it)
train_df["year"] = train_df["title"].str.extract(r"\((\d{4})\)")

# Step 2: Now safely remove the year from title
train_df["title"] = train_df["title"].str.replace(r"\(\d{4}\)", "", regex=True).str.strip()

# Optional: Convert year to numeric (in case there are any non-year cases)
train_df["year"] = pd.to_numeric(train_df["year"], errors="coerce")
# train_df["year"] = train_df["year"].astype("Int64")

train_df["year"] = pd.to_numeric(train_df["year"], errors="coerce").round().astype("Int64")
train_df
```

```python
# Popularity of genre

# Get the movies which have year specified
train_df_year_genre = train_df.dropna(subset=["year","genre"])

# Group by year and genre
genre_trend = train_df_year_genre.groupby(["year","genre"]).size().reset_index(name="count")

# plot a line chart for top genres, first pick top 
top_genres = train_df["genre"].value_counts().nlargest(6).index
filtered_trend = genre_trend[genre_trend["genre"].isin(top_genres)]
filtered_trend["year"] = filtered_trend["year"].astype("float64")
# Plot with seaborn
plt.figure(figsize=(14,7))
sns.lineplot(data=filtered_trend,x="year",y="count",hue="genre",marker="o")

plt.title("Genre Popularity Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.legend(title="Genre")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# LinearSVC Baseline
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score,f1_score,classification_report

n_samples,n_features = X_train.shape
dual_flag = not (n_samples > n_features) # if n_samples > n_features => dual=False

svc = LinearSVC(
    C=1.0,
    class_weight="balanced",
    max_iter=5000,
    dual=dual_flag,
    random_state=42
)

svc.fit(X_train,y_train)
y_preds_svc = svc.predict(X_test)

print("=== LinearSVC ===")
print("Accuracy:",accuracy_score(y_test,y_preds_svc))
print("Macro F1:",f1_score(y_test,y_preds_svc,average="macro"))
print("\nClassification Report:\n",classification_report(y_test,y_preds_svc,zero_division=0))
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    "C": [0.01, 0.1, 1, 10, 100],          # regularization strength
    "class_weight": [None, "balanced"]     # handle imbalance
}
```

```python
svc = LinearSVC(max_iter=5000, dual=False)  # dual=False works better for small datasets

grid = GridSearchCV(
    estimator=svc,
    param_grid=param_grid,
    scoring="f1_macro",    # optimize for macro F1
    cv=5,                  # 5-fold cross-validation
    n_jobs=-1,             # use all CPU cores
    verbose=2
)
```

```python
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Macro F1:", grid.best_score_)

best_model = grid.best_estimator_
```

```python
from sklearn.metrics import classification_report

y_pred = best_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define parameter grid for Logistic Regression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],   # Regularization strength
    'penalty': ['l2'],              # We'll keep L2 first (ridge)
    'solver': ['lbfgs']             # Best solver for multiclass + L2
}

# GridSearchCV setup
log_reg = LogisticRegression(max_iter=500, multi_class='multinomial')
grid_search = GridSearchCV(log_reg, param_grid, cv=3, scoring='accuracy', n_jobs=-1, verbose=2)

# Fit on training data
grid_search.fit(X_train, y_train)

print("Best Params:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)

# Evaluate on test set
best_log_reg = grid_search.best_estimator_
y_pred = best_log_reg.predict(X_test)

from sklearn.metrics import classification_report, accuracy_score
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
